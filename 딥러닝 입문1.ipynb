```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

```python
print(diabetes.data.shape, diabetes.target.shape)
```

```python
diabetes.data[0:3]
```

```python
diabetes.target[0:3]
```

```python
plt.scatter(diabetes.data[:,2],diabetes.target)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = diabetes.data[:,2]
y = diabetes.target
```

```python
w = 1.0
b = 1.0
```

```python
y_hat = x[0]*w+b
print(y_hat)
```

```python
print(y[0])
```

```python
w_inc = w + 0.1
y_hat_inc = x[0] * w_inc + b
print(y_hat_inc)
```

```python
w_rate = (y_hat_inc - y_hat) / (w_inc - w)
print(w_rate)
```
